```python
#create 3 equal chunks of list
l1=[2,4,6,8,9,1,3,5,7,6,0,1]
n=len(l1)//3
c1=l1[:n]
c2=l1[n:(n+n)]
c3=l1[(n+n):]
print(f"n: {n}\nc1: {c1}\nc2: {c2}\nc3: {c3}")
```

when u want to update anything in the tuple we should convert the tuple into list and then perform the required operations then again convert it back into the tuple. Because tuple is immutable in nature we cant perform the insert, update and delete operations directly.

```python
l1=[1,2,3,4,5]
l2=[0,9,8,7,6]
t1=()
for i in range(0,len(l1),1):
    t1=t1+((l1[i],l2[i]),)
t1
```

```python
slate="KavyaISAGoodGIRL@#123@gmaIL.COM"
#count number of upper, lower,special characters and numbers
uc=0
lc=0
sc=0
nd=0
for i in slate:
    if i.isupper():
        uc+=1
    elif i.islower():
        lc+=1
    elif i.isdigit():
        nd+=1
    else:
        sc+=1
print(f"Number of Upper:{uc}\nNumber of Lower:{lc}\nNumber of special characters:{sc}\nNumber of digits or numbers:{nd}")
```
