```python
#print prime numberrs between 1 to 20
for i in range(2,21,1):
    for j in range(2,i,1):
        if (i%j==0):
            break
    else:
        print(i, end=" ")
```

```python
#1) function with input and without return
def prime1(stv,spv):
   for i in range(stv,spv+1,1):
       for j in range(2,i,1):
           if (i%j==0):
               break
       else:
           print(i, end=" ")
```

```python
prime1(2, 30)
```

```python
#2) function with input and with return
def prime2(stv,spv):
    n=[]
    for i in range(stv,spv+1,1):
        for j in range(2,i,1):
            if (i%j==0):
                break
        else:
            n.append(i)
    return (n)
```

```python
prime2(2,34) #function already reached return so it is only printing the first prime number  
#now after adding the append list it is printing all the values we are storing the values in al list before it reaches to the return when all the values are done then at last it reaches the return and prints the list which gives the complete list of values
```

```python
#create a function to calculate overall sum and odd number sum, even number sum of the given range
#using with input and with return
a=int(input("Enter a value for a:"))
b=int(input("Enter a value for b:"))
total_sum=0
total_evensum=0
total_oddsum=0
for i in range(a,b+1,1):
    total_sum=total_sum+i
    if(i%2==0):
        total_evensum=total_evensum+i
    else:
        total_oddsum=total_oddsum+i
print(f"Total sum of the given numbers: {total_sum}\nTotal sum of the even numbers: {total_evensum}\nTotal sum of the odd numbers: {total_oddsum}")
```
