What is *Data Visualization?*
It is the process of representing data in a graphical or visual format, such as charts, graphs, or maps, so that people can easily understand patterns, trends, and insights. 

Why is visualization important?
* Makes data easy to understand
* Helps in quick decision-making
* Identifies patterns and trends
* detects outliers or errors
* improves communication of insights

**Univariate Analysis**
* It means analyzing only one variable at a time.
* To check the quality of the data
* Purpose:
* * Understand distribution
  * Find Patterns
  * Detect Outliers
  * Eg, Histogram, Boxplot, KDE Plot, Violin Plot

* Ex: Analyze salary distribution in a company -> are most employees low paid or high paid

**Bivariate Analysis**
* It means analyzing the relationship between two variable
* Purpose:
* * Understand the relationship
  * Identify the correlation
  * Compare variables
  * Eg: Scatter plot, bar plot, Line plot
* Ex: Analyze advertising spend vs Sales. Does increasing ads increase sales?

**Multivariate Analysis**
* It means analyzing the relationship between more than two variable
* Purpose:
* * Understand complex relationships
  * Identify patterns across multiple variables
  * Perform deep data analysis(EDA)
  * Eg: Pair plot, Heatmap, Facet grid, Bubble plot
* Ex: Analyze sales based on region, time, and product category -> which combination gives highest profit?

```python
#matplotlib and seaborn

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
#from matplotlib import pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
df=sns.load_dataset('tips')
df
```

```python
df.shape
```

**Histogram**
* Shows distribution of data
* Helps to detect skewness and spread

```python
plt.figure(figsize=(5,5))
plt.hist(df['total_bill'])
plt.title("Histogram")
plt.xlabel("Total bill")
plt.ylabel("Frequency")
plt.show() #Right Skew distribution, Mean>median
```

**KDE Plot** (Kernel Density Estimation Plot):
* Smooth version of histogram
* Better for understanding data density

```python
plt.figure(figsize=(5,5))
sns.kdeplot(df['total_bill'],fill=True)
plt.title("KDE Plot")
plt.xlabel("Total bill")
plt.ylabel("Frequency")
plt.show()
```

**Box Plot**
* Identifies outliers
* shows median,quartiles, spread

```python
plt.figure()
sns.boxplot(df['total_bill'],orient="h")
plt.title("Box Plot")
plt.xlabel("Total bill")
plt.show()
```

**Violin Plot**
* Combines box plot + distribution
* shows data shape clearly

```python
plt.figure()
sns.violinplot(x= df['total_bill'])
plt.title("Violin plot")
plt.show()
```

**Bar Plot**
* Compares values across categories
* Easy to understand differences

```python
plt.figure()
sns.barplot(x="day",y="total_bill",data=df)
plt.title("Bar Plot")
plt.show()
```

**Pie Chart**
* Shows proportion/percentage
* Good for quick insights
* note: use only for small num of categories

```python
plt.figure()
df['day'].value_counts().plot.pie(autopct='%1.2f%%')
plt.title('Pie Chart')
plt.show()
```

**Line plot**
* shows trends over time
* helps to identify increases, decreases, seasonality
* eg: sales growth month-wise.

```python
plt.plot(df['total_bill'])
plt.title('Line Plot')
plt.show()        
```

**Count Plot**
* shows frequency of categories
* useful for categorical analysis
* eg: number of customers per city

```python
sns.countplot(x='day',data=df)
plt.title('Count Plot')
plt.show()
```

```python
df['day'].value_counts()
```

**Scatter Plot**
* shows relationship between two numerical variables
* helps detect correlation and patterns
* eg: Advertising vs Sales

```python
sns.scatterplot(x='total_bill', y='tip',data=df)
plt.title("Scatter Plot")
plt.show()
```

**Bubble Sort**
* Add third variable(size)
* shows more information in one chart
* eg: sales vs profit with market size

```python
plt.scatter(df['total_bill'],df['tip'],s=df['size']*10)
plt.title("Bubble Sort")
plt.show()
```

**Pair Plot**
* Shows relationship between all variables
* useful in EDA (Exploratory data analysis)
* Eg: Feature relationships in ML

```python
sns.pairplot(df)
plt.show()
```

**Heat Map**
* Shows correlation between variables
* Helps in feature selection
* Eg: Identifies highly correlated features

```python
corr=df.corr(numeric_only=True)
corr.round(2)
```

```python
sns.heatmap(corr,annot=True)
plt.show()  # if we want to create a heatmap first, we should create a correlation
```
