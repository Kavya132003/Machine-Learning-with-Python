To extract a particular **row** and **column** we have 2 functions:
1. iloc: index location
2. loc: location name

```python
iloc syntax:

var.iloc[stp row:spp row:ss row, stp column: spp column: ss column] 

default values:
start posistion(stp): 0
stop position(spp): upto last position
Step size(ss): 1
```

```python
cars
```

```python
#all the details of first 5 rows with all the columns
cars.iloc[0:5:1,0::1]
```

```python
#all the rows with first 5 columns
cars.iloc[0::1,0:5:1]
```

```python
#same above question
cars.iloc[:,:5]
```

```python
#extract last 5 rows with last 5 columns
cars.iloc[-5:,-5:]
```

```python
#extract even index rows with odd index columns
cars.iloc[::2,1::2]
```

```python
#extract index whose rows are divisible by 4 and columns from 3rd to 7th 
cars.iloc[4::4,2:7:1]
```

```python
#extract some random rows and columns with there column numbers rows[1,4,6,3,37,3,8] and columns[3,5,10,2,6]
cars.iloc[[1,4,6,3,7,3,8],[3,5,10,2,6]]
```

```python
loc syntax:

var.loc[stn row:spn row:step size, stn column:spn column:step size]
```

```python
cars
```

```python
#extract 1st 5 rows with 1st 5 columns
cars.loc[0:4,'manufacturer':'hp']
```

```python
#reverse the table
cars.loc[31::-1,'carb'::-1]
```

```python
#import insurance dataset
insu=pd.read_csv("File Path")
insu
```

```python
"""On insurance data 
1) each region wise total expenses
2) Gender wise average BMI and expenses
3) Each region, each children and smoker class wise total expenses 

Use: Groupby and pivot table"""
```

```python
#1) each region wise total expenses
insu.groupby(by='region')['expenses'].sum().sort_values()(asecending=False)
```

```python
pd.pivot_table(data=insu, index='region',values='expenses',aggfunc='sum')
```

```python
insu.groupby(by='sex')[['bmi','expenses']].mean().round(3)
```

```python
pd.pivot_table(data=insu,index='sex',values=['bmi','expenses'],aggfunc='mean').round(3)
```

```python
insu.groupby(by=['region','children','smoker']).['expenses'].mean().round(3)
```

```python
pd.pivot_table(data=insu,index='region',columns=['smoker','children'],values='expenses',aggfunc='mean').round(3)
```
