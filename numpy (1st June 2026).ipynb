**What is NumPy?**
It is a powerful Python library used for 
* Numerical computations
* Working with arrays and matrices
* Mathematical operations
* Data Science and Machine Learning

NumPy: Numerical Python

Advantages of NumPy:
* Faster than Python lists
* Uses less memory
* Supports multi-dimensional arrays
* Large collections of mathematical functions

**List vs Array:**

1. List we can not do element wise operations, but in an array we can.
2. Array to list conversion is possible, and also list to array conversion is possible, but it will lose the dimension.
3. List is heterogenous datatype, but NumPy is a homogeneous data type.

```python
a=np.array([1,2,3,4,5])
b=np.array([4,5,6,7,8])
```

```python
#extract common elements from both the arrays
np.intersect1d(a,b)
```

```python
#find elements present in array a but not in array b
np.setdiff1d(a,b)
```

```python
#find elements present in array b but not in array a
np.setdiff1d(b,a)
```
