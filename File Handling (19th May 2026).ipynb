**Error (also called exceptions):**
* Errors are the issues that stop your program from running properly.
* Error handling means detecting and gracefully managing errors instead of letting the program crash.

1) Syntax error
2) Name error
3) Indent error
4) Type error
5) zerodivision error
6) index error
7) value error
8) key error
9) filenotfound error



**syntax of exception:**

try:
    #code that might cause an error
except:
    #code to handle the error

```python
x=int(input("Enter a value:"))
print(10/x)
```

```python
try:
    x=int(input("Enter a value:"))
    print(10/x)
except:
    print("The x should not be zero, If it is zero it throws the zerodivision error")
```

```python
try:
    a=int(input("Enter a value for a:"))
    b=int(input("Enter a value for b:"))
    c=a/b
    print("Answer:",c)
except ValueError:
    print("The entered value is wrong, Please enter the valid values")
except ZeroDivisionError:
    print("The value of b cant be zero, Please enter some other number")
except:
    print("Something went wrong")
```

```python
try:
    a=int(input("Enter a value for a:"))
    b=int(input("Enter a value for b:"))
    c=a/b
    print("Answer:",c)
except ValueError:
    print("The entered value is wrong, Please enter the valid values")
except ZeroDivisionError:
    print("The value of b cant be zero, Please enter some other number")
except:
    print("Something went wrong")
```

```python
try:
    a=int(input("Enter a value for a:"))
    b=int(input("Enter a value for b:"))
    c=a/b
    print("Answer:",c)
except (ValueError, ZeroDivisionError):
    print("The entered value is wrong, Please enter the valid values")
```
