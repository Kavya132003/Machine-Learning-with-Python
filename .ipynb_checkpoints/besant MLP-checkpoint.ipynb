```python
#21st April 2026
DataTypes: It defines about the type of data we storing to the variable
there are 2 types of data type

1) SingleValued datatypes
int, float, str(always store in double qoutes""), bool, complex

2) MultiValued datatypes
list, tuple, dict,set

OPERATORS: It is a special symbol which performs certain operations between two operands is called as operator.
eg: 3+6
3,6-> operands      +-> operators

There are 7 different types of operators in python

1) Arithematic Operators: + - * / %(it gives remainder) //(it gives quotient value, floor division) **(exponential value)
2) Assignment Operators: = += -= *= /= %= //= **=
3) Logical Operators: and or not
4) Relational or Comaprison Operators: == != > < >= <=
5) Bit Wise Operators: & | ~ >> <<
6) Membership Operators: in not in (to check the presence)
7) Identity Operators: is is not
```

```python
Conditional Statements: 
They are used to make decisions in program based on conditions
You can execute only one set of statements either if or else
Different types of Conditional Statements:
1) if
2) if else
3) continous if else/ if else ladder
4) nested if else

1)if
syntax: 
if condition:
    statements

2)if else
syntax:
if condition:
    statements
else:
    statements
```

```python
#check the person is adult or not
age=int(input("Enter the age of a person:"))
if age>=18:
    print("The Person is adult")
else: 
    print("The Person is minor")
```

```python
#write a program to check the given num is odd or even
num=int(input("Enter any value:"))
if num%2==0:
    print("The number is even")
else:
    print("The number is odd")
```

```python
27th April 2026

Nested Conditional Statements
syntax:
if cond1: #outer if condition
    if cond2: #inner if condition
        statements of cond2
    else: #inner else
        statements of inner else
else: #outer else
    statements of outer else

```

```python
n = int(input("Enter a number:"))
if n<=0:
    if n==0:
        print("The given number is zero")
    else:
        print("The Given number is negative")
else:
    print("The given number is positive")
```

```python
#temperature checker
#<=15, moderate; >15, cold; >30, hot; >40

temperature = float(input("Enter the temperature value:"))
if temperature<=30:
    if temperature>15:
        print("The temperature is cold")
    else:
        print("The temperature is moderate")
elif temperature<40:
    print("The temperture is hot")
else: 
    print("The temperature is very hot")
            
```
