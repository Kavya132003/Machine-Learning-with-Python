6) Membership Operators: in not in (to check the presence)
7) Identity Operators: is is not

```python
"""Based on percentage print the result
85% - 100% - Distinction
60% - 84% - 1st class
50% - 59% - 2nd class
>35% - 49% - just pass
<=35% - Fail"""

Percentage=float(input("Enter the percentage gained by the student:"))
if (Percentage<0 or Percentage>100):
    print("Invalid Input")
elif (Percentage>=85):
    print("Distinction")
elif (Percentage>=60):
    print("1st Class")
elif (Percentage>=50):
    print("2nd Class")
elif (Percentage>35):
    print("Just pass")
else:
    print("Fail")
```

```python
"""Check the given inputs that they form which type of triangle 
Equilateral
Isosceles
Scalene"""

x= float(input("Enter a value for x:"))
y= float(input("Enter a value for y:"))
z= float(input("Enter a value for z:"))
if (x==y==z):
    print("Equilateral")
elif (x!=y!=z!=x):
    print("Scalene")
else:
    print("Isosceles")
```

```python
#Assignment2

""" 1. E-commerce Discount Calculator

Scenario: An online store offers discounts based on the purchase amount:
10% discount for purchases between $100 and $500
20% discount for purchases above $500
No discount for purchases below $100
Task: Write a program that takes the purchase amount as input and calculates the discount and final amount to be paid. """

amount = float(input("Enter the purchase amount: "))
if amount > 500:
    discount = amount * 0.20
elif amount >= 100:
    discount = amount * 0.10
else:
    discount = 0
final_amount = amount - discount
print("Discount Amount =", discount)
print("Final Amount to Pay =", final_amount)
```

```python
"""2. Traffic Light Simulation

Scenario: Create a program that simulates a traffic light. The user inputs one of the colors: Red, Yellow, or Green. Based on the input:
If the input is Red, display "Stop."
If the input is Yellow, display "Ready to move."
If the input is Green, display "Go."
For invalid input, display "Invalid color." """

color = input("Enter the traffic light color: ")
if color.lower() == "red":
    print("Stop")
elif color.lower() == "yellow":
    print("Ready to move")
elif color.lower() == "green":
    print("Go")
else:
    print("Invalid color")
```

```python
""" 8. Temperature Alert System

Scenario: A weather monitoring system alerts based on the temperature:
Below 0°C: "Freezing weather"
0°C to 20°C: "Cold weather"
21°C to 30°C: "Warm weather"
Above 30°C: "Hot weather"
Task: Write a program that takes the temperature as input and displays the corresponding alert message """

temperature = float(input("Enter the temperature in Celsius: "))
if temperature < 0:
    print("Freezing weather")
else:
    if temperature <= 20:
        print("Cold weather")
    else:
        if temperature <= 30:
            print("Warm weather")
        else:
            print("Hot weather")
```

```python
"""5. ATM Withdrawal

Withdrawal amount ≤ balance → Proceed
Withdrawal amount must be multiple of 100
Otherwise → Transaction not allowed"""

balance = int(input("Enter account balance: "))
withdrawal_amount = int(input("Enter withdrawal amount: "))

if withdrawal_amount <= balance:
    if withdrawal_amount % 100 == 0:
        print("Transaction successful")
    else:
        print("Transaction not allowed: Amount must be a multiple of 100")
else:
    print("Transaction not allowed: Insufficient balance")
```

```python
"""7. Electricity Bill

Units ≤ 50 → ₹2/unit
Units > 50 and ≤ 100 → ₹3/unit
Units > 100 and ≤ 200 → ₹5/unit
Units > 200 → ₹8/unit"""

units = int(input("Enter the number of units consumed: "))
if units <= 50:
    bill = units * 2
elif units <= 100:
    bill = units * 3
elif units <= 200:
    bill = units * 5
else:
    bill = units * 8
print("Electricity Bill = ₹", bill)
```

```python
"""
8. Shopping Discount

Amount < 500 → No discount
Amount ≥ 500 and < 1000 → 5% discount
Amount ≥ 1000 and < 5000 → 10% discount
Amount ≥ 5000 → 20% discount"""

amount = float(input("Enter shopping amount: "))
if amount < 500:
    discount = 0
elif amount < 1000:
    discount = amount * 0.05
elif amount < 5000:
    discount = amount * 0.10
else:
    discount = amount * 0.20
final_amount = amount - discount
print("Discount = ₹", discount)
print("Amount to Pay = ₹", final_amount)
```

```python
"""9. Loan Eligibility

Salary ≥ 25000 and credit score ≥ 750 → Approved
Salary ≥ 25000 and credit score < 750 → Review
Salary < 25000 and credit score ≥ 800 → Approved
Otherwise → Rejected"""

salary = int(input("Enter your salary: "))
credit_score = int(input("Enter your credit score: "))
if salary >= 25000 and credit_score >= 750:
    print("Loan Approved")
elif salary >= 25000 and credit_score < 750:
    print("Loan Under Review")
elif salary < 25000 and credit_score >= 800:
    print("Loan Approved")
else:
    print("Loan Rejected")
```

```python
"""10. Exam Result System

All subjects ≥ 35 → Pass
Any subject < 35 → Fail
If pass:
Average ≥ 85 → Distinction
Average ≥ 60 and < 85 → First Class
Average < 60 → Pass Class
If any subject < 20 → Fail (Severe)"""

sub1 = int(input("Enter marks in Subject 1: "))
sub2 = int(input("Enter marks in Subject 2: "))
sub3 = int(input("Enter marks in Subject 3: "))
average = (sub1 + sub2 + sub3) / 3
if sub1 < 20 or sub2 < 20 or sub3 < 20:
    print("Fail (Severe)")
elif sub1 < 35 or sub2 < 35 or sub3 < 35:
    print("Fail")
else:
    if average >= 85:
        print("Pass - Distinction")
    elif average >= 60:
        print("Pass - First Class")
    else:
        print("Pass - Pass Class")
```

```python
"""
11. Ride Fare Calculation

Distance ≤ 5 km and non-peak → ₹10/km
Distance ≤ 5 km and peak → ₹15/km
Distance > 5 km and non-peak → ₹8/km
Distance > 5 km and peak → ₹12/km"""

distance = float(input("Enter distance in km: "))
peak = input("Is it peak time? (yes/no): ")
if distance <= 5 and peak == "no":
    fare = distance * 10
elif distance <= 5 and peak == "yes":
    fare = distance * 15
elif distance > 5 and peak == "no":
    fare = distance * 8
else:
    fare = distance * 12
print("Total Fare = ₹", fare)
```

```python
"""13. Employee Bonus

Experience < 2 years → No bonus
Experience ≥ 2 and < 5 → 10% bonus
Experience ≥ 5 and salary < 50000 → 15% bonus
Experience ≥ 5 and salary ≥ 50000 → 20% bonus"""

salary = float(input("Enter salary: "))
experience = float(input("Enter years of experience: "))
if experience < 2:
    bonus = 0
elif experience < 5:
    bonus = salary * 0.10
elif salary < 50000:
    bonus = salary * 0.15
else:
    bonus = salary * 0.20
print("Bonus = ₹", bonus)
```

```python
"""14. Movie Ticket Pricing

Age < 12 → ₹100
Age ≥ 12 and < 60 → ₹200
Age ≥ 60 → ₹150
Evening show → Additional ₹50"""

age = int(input("Enter age: "))
show = input("Is it an evening show? (yes/no): ")
if age < 12:
    ticket_price = 100
elif age < 60:
    ticket_price = 200
else:
    ticket_price = 150
if show == "yes":
    ticket_price += 50
print("Ticket Price = ₹", ticket_price)
```

```python
"""15. Food Delivery Charges

Order < ₹500 → Delivery charge ₹50
Order ≥ ₹500 → Free delivery
Distance > 10 km → Extra ₹30 charge"""

order_amount = float(input("Enter order amount: ₹"))
distance = float(input("Enter delivery distance (km): "))
if order_amount < 500:
    delivery_charge = 50
else:
    delivery_charge = 0
if distance > 10:
    delivery_charge += 30
print("Delivery Charge = ₹", delivery_charge)
print("Total Amount = ₹", order_amount + delivery_charge)
```
