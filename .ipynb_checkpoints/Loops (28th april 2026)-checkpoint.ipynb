Loops:
These are used to repeat a block of code multiple times until a condition is met

Types of loops
1) for loop
    i) for loop with sequence
    ii) for loop with range
2) while loop
3) nested loop
    i) for loop inside for loop
    ii) for loop inside while loop
    iii) while loop inside while loop
    iv) while loop inside for loop

1) for loop
    i) for loop with sequence
its used iterate the sequence data 

syntax:
for itt_var in sequence_name:
    statements

Entire loop consists of these 3 things:
* initialization
* condition
* incrementation/decrementation

Python for loop has a condition called range.

for itt_var in range(start_value(initialization) ,stop_value(condition) ,step_size(increment/decrement value):
    statements

default start value is 0
default step size is 1

```python
#write a program to print sum of digits of the given number
n=int(input("Enter a number:"))
total=0
while(n>0):
    i=n%10
    total=total+i
    n=n//10
print("The total sum of digits is:",total)
```

```python
# Palindrome sequence
# Write a program to check whether the given number is a palindrome or not 

n=int(input("Enter any number:"))
temp=n
rev=0
while (n>0):
    rem = n%10
    rev = (rev*10)+rem
    n=n//10
print("Original n=", temp)
print("Reverse:", rev)
if (rev==temp):
    print("The number given is palindrome")
else:
    print("The number given is not palindrome")
```

i) for loop inside for loop
syntax:

for itt_var in range(stv,spv,ss): #outer loop
    for itt_var in range(stv,spv,ss): #inner loop
        inner loop statements
    outer loop statements

for 1 outer loop value the inner loop will complete its cycle


```python
for i in range (1, 9, 1):
    for j in range(1, 6, 1):
        print(j, end="\t")
    print("\n")
```

syntax:

for itt_var in range(stv,spv,ss): #outer loop
    initialization of while
    while (cond):
        statements of while
        inc/dec of while
    statements of for loop

iii) while loop inside while loop
syntax:

initialization of outer while
while(cond): #outer while
    initialization of inner while
    while(cond): #inner while
        statements of inner while
        inc/dec of inner while
    statements of outer while
    inc/dec of outer while

```python
i=1
while (j<=5):
    j=1
    while(j<=i):
        if (i==1 or j==1 or i==5 or j==i):
            print("*", end=" ")
        else:
            print(" ", end=" ")
        j+=1
    i+=1
    print()
```

```python
#upper body
i=1
while (i<=5):
    for j in range(5,i-1,-1):
        print("*", end=" ")
    print()
    i+=1

#lower body
i=2
while (i<=5):
    for j in range(1,i+1,1):
        print("*", end=" ")
    print()
    i+=1
```

```python
for i in range(1,6,1):
    for j in range(4,i-1,-1):
        print(" ", end="")
    for k in range(1,i+1,1):
        print("*",end=" ")
    print()
```

```python
rows = 5

for i in range(1, rows + 1):
    for j in range(rows - i):
        print(" ", end="")
    for k in range(i):
        print("* ", end="")
    print()

for i in range(rows - 1, 0, -1):
    for j in range(rows - i):
        print(" ", end="")
    for k in range(i):
        print("* ", end="")
    print()
```
