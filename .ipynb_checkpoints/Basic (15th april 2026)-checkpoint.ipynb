2)Write about the history of Python?
->History of Python
Python was created by Guido van Rossum in the late 1980s at Centrum Wiskunde & Informatica (CWI) in the Netherlands.
The language was designed as a successor to the ABC programming language.
Python was officially released in 1991 with a focus on simplicity and readability.
The name “Python” was inspired by the British comedy show Monty Python's Flying Circus.
Python 2.0 was released in 2000 with new features like garbage collection and Unicode support.
In 2008, Python 3.0 was introduced with major improvements and better code consistency.
Python gained massive popularity in web development, automation, artificial intelligence, and data science.
Today, Python is maintained by the Python Software Foundation and is one of the world’s most widely used programming languages.

3)Explain different versions of Python?

-> Python 1.x
Python 1.0 was released in 1991.
It included basic features like functions, modules, and exception handling.
This version focused on simple and readable programming.

-> Python 2.x
Python 2.0 was released in 2000.
It introduced features like list comprehensions, garbage collection, and Unicode support.
Python 2.x became very popular for software and web development.
Official support for Python 2 ended in 2020.

-> Python 3.x
Python 3.0 was released in 2008 with major improvements.
It removed outdated features and improved code readability.
Better support for Unicode and modern programming practices was added.
Python 3.x is the current and most widely used version today.
Popular Python 3 Versions
Python 3.6 → Introduced formatted string literals (f-strings).
Python 3.7 → Improved performance and data classes support.
Python 3.8 → Added walrus operator (:=).
Python 3.9 → Included dictionary merge operators.
Python 3.10 → Introduced pattern matching feature.
Python 3.11 → Improved execution speed and error messages.
Python 3.12 → Enhanced performance and removed deprecated features.
