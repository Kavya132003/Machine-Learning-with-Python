```python
#calculate avg,max,min,variance,standard deviation and total of mpg column

#Maximum
cars[['mpg']].max()
```

```python
#Minimum
cars[['mpg']].min()
```

```python
#Average or mean
cars[['mpg']].mean()
```

```python
#Variance
cars[['mpg']].var()
```

```python
#standard deviation
cars[['mpg']].std()
```

```python
#sum or total
cars['mpg'].sum()
```

```python
cars['cyl'].value_counts() #total count of unique values 
```

```python
#extract the records of 6 unique values of cyl cars along with there complete details
cars[cars['cyl']==6]
```

```python
#extract the records whose hp is between 100 to 200
cars[(cars['hp']>=100) & (cars['hp']<=200)]
```

```python
#extract the records whose hp is not between 100 to 200
cars[(cars['hp']<100) | (cars['hp']>200)]
```

```python
#extract cars with 4 gears and 1 carb
cars[(cars['gear']==4) & (cars['carb']==1)]
```

```python
#extract cars with disp between 100 to 200 with 5 gears
cars[(cars['disp']>=100) & (cars['disp']<=200)&(cars['gear']==5)]
```

```python
#extract cars with either 10 to 20 qsec or vs should be 0
cars[((cars['qsec']>=10) & (cars['qsec']<=20))|(cars['vs']==0)]
```

```python
#display each gear-wise maximum mileage
cars.groupby(by='gear')['mpg'].max()
```
