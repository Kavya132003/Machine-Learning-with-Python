```python
#28th April 2026
Loops:
These are used to repeat a block of code multiple times until a condition is met

Types of loops
1) for loop
    i) for loop with sequence
    ii) for loop with range
2) while loop
3) nested loop
    i) for loop inside for loop
    ii) for loop inside while loop
    iii) while loop inside while loop
    iv) while loop inside for loop

1) for loop
    i) for loop with sequence
its used iterate the sequence data 
syntax:
for itt_var in sequence_name:
    statements

initialization
condition
incrementation/decrementation

for itt_var in range(start_value,stop_value,step_size):
    statements

default start value is 0
default step size is 1
```

```python
#without function
name= "Kavya"
p=0
for i in name:
    print(p,i)
    p+=1
```

```python
#write a program to calculate the factorial of a given number
num=int(input("Enter a number:"))
factorial=1
for i in range(1,num+1,1):
    factorial*=i
print("The factorial of a given number is:",factorial)
```

```python
#print numbers between 1 to 50 which are divisible by both 3 and 5
for i in range(1,51,1):
    if i%3==0 and i%5==0:
        print(i,end=" ")
```

```python
#write a program to check the given number is prime or not
num=int(input("Enter a number:"))
count=0
for i in range(2,num):
    if num%i==0:
        count=count+1
print("count=", count)
if count==0:
    print(num,"is prime number")
else:
    print(num,"is not a prime number")
```

```python
#write a program to print sum of digits of the given number
n=int(input("Enter a number:"))
total=0
while(n>0):
    i=n%10
    total=total+i
    n=n//10
print("The total sum of digits is:",total)
```
